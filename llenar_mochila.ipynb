Llenar una mochila: Crea una funcin que reciba una lista de pesos de objetos y un peso mximo, y determine cuntos objetos puedes llevar sin superar el peso mximo. El objetivo es seleccionar la mayor cantidad de objetos posible sin que la suma de sus pesos exceda el lmite permitido. Las reglas son:
- Los objetos deben seleccionarse de manera que el peso total no supere el peso mximo.
- Debes intentar incluir el mayor nmero posible de objetos.
- La funcin debe devolver el nmero total de objetos que se pueden llevar sin exceder el peso mximo.
- Los pesos de los objetos son nmeros enteros positivos.
- Los objetos deben seleccionarse en el orden que maximice la cantidad de objetos, por lo que es til ordenar los objetos por peso ascendente antes de seleccionarlos.



```python
def llenar_mochila(pesos, peso_max):

    copia_pesos = list(pesos)
    suma_peso = 0
    cantidad_objetos = 0
    while len(copia_pesos) > 0:
        for i,peso in enumerate(copia_pesos):
            suma_peso += peso
            print(f"I: {i}")
            print(f"peso: {peso}")
            print(f"peso total: {suma_peso}")

            if suma_peso > peso_max:
                suma_peso -= peso
                copia_pesos.pop(copia_pesos.index(peso))
            else:
                cantidad_objetos += 1
                print(f"Cantidad obje: {cantidad_objetos} \n")

        print(cantidad_objetos)
        


llenar_mochila([1,29,16], 179)
```
