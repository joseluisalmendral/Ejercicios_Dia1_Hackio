Encontrar el nmero que falta en una secuencia:  Crea una funcin que reciba una lista de n-1 nmeros nicos que representan una secuencia de nmeros consecutivos entre 1 y n, donde falta un nmero, y encuentre el nmero faltante. El objetivo es identificar el nmero que falta para completar la secuencia. Las reglas son:
- La secuencia contiene nmeros consecutivos entre 1 y n, con exactamente un nmero faltante.
- La lista contiene n-1 nmeros nicos.
- La funcin debe devolver el nmero faltante.
- La funcin debe ser capaz de manejar tanto secuencias ascendentes como descendentes, siempre que contengan nmeros consecutivos.
- El valor de n es el nmero ms alto esperado en la secuencia.


```python
def findMissingNumbers(n):
    numbers = sorted(list(set(n)))
    output = []
    """Funcin que identifica nmeros faltantes en una lista

    Returns:
        Nmero mximo que falta
    """
    for i in range(numbers[0], n[-1]):
        if i not in numbers:
            output.append(i)
    return max(output)

listOfNumbers = [1101,1102,1104]
print(findMissingNumbers(listOfNumbers))
```
